```python
import pandas as pd
pd.options.mode.chained_assignment=None
```

```python
covid=pd.read_csv('us-states.csv')
```

```python
covid.shape
```

```python
covid.head()
```

```python
state='Massachusetts'
covid_state=covid[covid.state==state]
covid_state['daily_cases']=covid_state['cases']-covid_state['cases'].shift(1)
covid_state['daily_deaths']=covid_state['deaths']-covid_state['deaths'].shift(1)
covid_state=covid_state.fillna(0)
```

```python
covid_state.tail(20)
```

```python

covid_state.daily_deaths.plot.bar()
```

```python
reverse=covid_state.iloc[::-1]
```

```python
markdown=reverse.to_markdown();
```

```python
fname=state+".md"
fname=fname.replace(" ","_")
print(fname)
```

```python
out_file=open(fname, "w")
out_file.write(markdown)
out_file.close()
```
